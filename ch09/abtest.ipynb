## 5. ABTEST 기초개념

### 기본 단계
___
- 목표 설정
    - 테스트의 목적을 명확히 하고, 무엇을 측정할지 결정. 

    - (ex) 웹페이지에서의 클릭률, 구매 전환율, 또는 사용자의 체류 시간 등

- 가설 수립
    - 일반적으로 "버전 B는 버전 A보다 더 나은 성능을 보일 것이다"와 같은 가설을 설정. 

    - 이 가설은 테스트의 결과를 예측하고, 이를 검증하는 데 사용됨.

- 대상 그룹 분할
    - 사용자를 무작위로 두 그룹(A와 B)으로 나누어 각 그룹에게 다른 버전을 제공. 
    - 이렇게 함으로써 외부 요인의 영향을 최소화하고 두 버전의 차이가 실제 효과 때문인지를 판단하게됨.

- 실험 실행
    - 두 버전을 동시에 실행하여 데이터를 수집.

    - 이 단계는 충분한 데이터를 확보할 때까지 계속됨.

- 데이터 분석
    - 수집된 데이터를 분석하여 어느 버전이 더 나은 성능을 보이는지 통계적으로 검증함. 

    - t-검정, z-검정 등의 통계 검정 방법을 사용할 수 있음.


### 실험 변수의 타입에 따른 차이
___
- 이산형 변수
    - 실험 변수가 카테고리형 데이터를 가지고 있음.

    - (ex) 구매여부, 클릭여부, 이탈여부, 재방문여부
    - 통계 검정 방법 : 두 비율 검정, 카이제곱 검정, Fisher 정확 검정(Fisher's Exact Test)


```python
import numpy as np
import statsmodels.stats.power as smp
from statsmodels.stats.proportion import proportion_effectsize

# 가정: 현재 전환율이 2%, 목표 전환율을 2.5%로 설정하여 25% 증가를 목표로 함
# 효과 크기를 계산 (Cohen's h 사용)
# Cohen's h = 2 * arcsin(sqrt(p1)) - 2 * arcsin(sqrt(p2))
current_rate = 0.02  # 현재 전환율
target_rate = 0.025  # 목표 전환율
effect_size = 2 * (np.arcsin(np.sqrt(target_rate)) - np.arcsin(np.sqrt(current_rate)))
# 라이브러리 사용 : effect_size = proportion_effectsize(current_rate, target_rate) 

# 샘플 사이즈 추정을 위한 검정력 계산 (목표 검정력: 0.8, 유의 수준: 0.05)
alpha = 0.05
power = 0.8
sample_size = smp.NormalIndPower().solve_power(effect_size, power=power, alpha=alpha, ratio=1, alternative='two-sided')

# 결과 출력
print(f"계산된 효과 크기: {effect_size:.4f}")
print(f"필요한 최소 표본 크기 (각 그룹): {sample_size:.0f}")

```

> 프롬프트 : effect size는 무엇인가요?

- 연속형 변수
  - 실험 변수가 수치형 데이터를 가지고 있음.

  - (ex) 구매 금액, 체류 시간, 객단가
  - 통계 검정 방법 : t-검정, ANOVA, MANOVA

```python
import statsmodels.stats.power as smp
import numpy as np

# 평균 구매 금액, 목표 구매 금액, 가정의 표준 편차
current_mean = 50  # 현재 평균 구매 금액
target_mean = 55  # 목표 구매 금액
std_dev = 10  # 가정된 표준 편차

# Cohen's d 효과 크기 계산
effect_size = (target_mean - current_mean) / std_dev

# 샘플 사이즈 계산
alpha = 0.05  # 유의 수준
power = 0.8  # 검정력
sample_size = smp.TTestIndPower().solve_power(effect_size=effect_size, alpha=alpha, power=power, ratio=1.0, alternative='larger')

# 결과 출력
print(f"계산된 효과 크기: {effect_size:.4f}")
print(f"필요한 최소 표본 크기 (각 그룹): {sample_size:.0f}")
```

### 중심극한의 정리 (Central Limit Theorem)와 대수의 법칙 (Law of Large Numbers)

1.	중심극한정리 (CLT): 충분히 큰 샘플사이즈가 있는 경우, 어떤 분포든지 표본평균은 정규분포를 따르게 됩니다.
- CLT: 이렇게 모은 평균들이 → 통계적으로 분석하기 좋게 정규분포처럼 행동하게 됨
2.	대수의 법칙 (LLN): 샘플사이즈가 커질수록 표본평균이 실제 기대값에 가까워지게 됩니다.
- LLN: 테스트 데이터를 많이 모으면 → 우리 결과는 **진짜 효과(전환율 등)**에 근접함

> 그래서 우리는 p-value 계산, 신뢰구간 추정 등을 정규분포 기반으로 안심하고 할 수 있음

### ❓실전 Online ABTest 분석가 GPTs 에게 질문 (접속링크 : https://bit.ly/4ksE910)
> 프롬프트 : 중심극한정리와 대수의 법칙을 각각 설명해주세요.

```python
import numpy as np
import plotly.graph_objs as go
from plotly.subplots import make_subplots
from scipy.stats import norm

# 설정
p = 0.3  # 구매 확률
sample_sizes = [10, 50, 100, 500, 1000]  # 샘플사이즈 목록

# 이항분포 시뮬레이션 함수
def simulate_binomial(n, p, size):
    return np.random.binomial(n, p, size)

# 슬라이더를 사용하여 상호작용 가능한 플롯 생성
fig = make_subplots(rows=1, cols=1)

# 처음에는 첫 번째 샘플사이즈로 히스토그램과 정규분포 곡선 생성
initial_n = sample_sizes[0]
binom_data = simulate_binomial(initial_n, p, 10000)
mean = initial_n * p
std_dev = np.sqrt(initial_n * p * (1 - p))
x = np.linspace(min(binom_data), max(binom_data), 1000)
norm_curve = norm.pdf(x, mean, std_dev)

# 이항분포 히스토그램
hist = go.Histogram(x=binom_data, histnorm='probability density', opacity=0.6, name='이항 분포')
# 정규분포 곡선
curve = go.Scatter(x=x, y=norm_curve, mode='lines', name='정규 근사')

fig.add_trace(hist)
fig.add_trace(curve)

# 슬라이더 설정
steps = []
for n in sample_sizes:
    binom_data = simulate_binomial(n, p, 10000)
    mean = n * p
    std_dev = np.sqrt(n * p * (1 - p))
    x = np.linspace(min(binom_data), max(binom_data), 1000)
    norm_curve = norm.pdf(x, mean, std_dev)
    
    step = dict(
        method='update',
        args=[{'x': [binom_data, x],
               'y': [hist['y'], norm_curve]},
              {'title': f'샘플사이즈 = {n}'}]
    )
    steps.append(step)

sliders = [dict(
    active=0,
    currentvalue={"prefix": "샘플사이즈: "},
    pad={"t": 50},
    steps=steps
)]

fig.update_layout(
    sliders=sliders,
    title=f'샘플사이즈 = {initial_n}',
    xaxis_title='구매횟수',
    yaxis_title='빈도',
)

fig.show()
```

### 최소 감지 효과(MDE)
___
- 효과 크기(Effect Size)
    - 두 집단 간의 차이나 상관관계의 크기를 수치적으로 표현

    - 즉, 실험의 결과가 실제로 얼마나 의미 있는지를 나타내는 척도
    - (ex) 두 치료법의 효과 차이, 교육 방법의 성과 차이 등
    - 일반적인 효과 크기 지표로는 Cohen's d, Pearson's r 등
    - 효과 크기가 크면 클수록, 실험 결과의 실제적 중요성이 높음

- 최소 감지 효과(MDE)
    - 유의미한 차이를 감지하기 위해 최소한 필요한 효과 크기를 의미

    - 샘플사이즈를 계산할때 활용함.
    
- MDE 설정 방법 (애매하고 주관적인 영역)
    - 실험의 목적과 문맥 파악

        - (ex) 전환율을 현재의 2%에서 2.5%로 25% 증가시키는 것이 비즈니스 목표일 수 있음.
    - 이전 데이터 참고
        - 현재상황의 평균 전환율을 분석하여 효과 크기의 초기 추정치를 설정
    - 리소스의 제한 고려
        - 예산, 시간, 사용 가능한 트래픽 양 등을 고려하여 현실적인 범위 내에서 결정
    - 기타 방법
        - 유사 산업 또는 경쟁사에서 보고된 전환율 개선 사례를 참고
        - 파일럿 테스트로 예상 효과 가늠
        - 통계적 자문
- 공식

### 유의수준
___

- 유의 수준(Significance Level, α)
    - 유의 수준 설정은 연구의 엄격함을 결정하는 중요한 요소

    - 통계적 가설 검정에서 귀무 가설을 잘못 기각할 위험을 감수하는 최대 확률
    - 즉, 유의 수준을 설정함으로써, 귀무 가설이 사실일 때 이를 잘못 기각할 가능성을 정의

- 유의 수준 설정 방법 (산업별 표준)
    - 일반적으로 0.05(5%) 또는 0.01(1%)로 설정

    - 유의 수준이 낮을수록 결과가 우연에 의한 것이 아니라는 확신이 높아짐
   
- 결론
    - 효과 크기와 유의 수준은 실험의 설계와 해석에서 중요하게 고려됨. 

    - 효과 크기는 실험의 실질적 중요성을, 
    - 유의 수준은 결과의 통계적 신뢰성을 각각 나타냄. 
    - 이 두 매개변수는 검정력과 밀접하게 연관됨


### 통계 검정력
___
- 정의:
    - 통계적 가설 검정에서 귀무 가설이 거짓일 때 이를 기각할 확률, 

    - 즉, 실험에서 실제 효과를 놓치지 않고 정확하게 감지하는 능력

- 중요성
    - 낮은 검정력: 실제 효과를 감지 못하고 귀무 가설을 잘못 채택할 위험 증가

    - 높은 검정력: 귀무 가설이 거짓일 때 이를 정확히 기각할 확률이 높아, 실제 차이를 더 잘 포착
    - 설계 단계에서 필요한 최소 표본 크기 추정하는데 활용

- 요소별 영향
    - 표본 크기(n): 표본 크기가 클수록 검정력 증가

    - 효과 크기(effect size): 효과 크기가 클수록 검정력 증가
    - 유의 수준(α, alpha): 유의 수준이 높을수록 검정력 증가하지만, 오류 발생 가능성도 증가

* <mark> MDE, 유의 수준, 검정력을 이용한 필요 샘플사이즈 계산</mark>

```python
import gradio as gr
from statsmodels.stats.power import TTestIndPower, NormalIndPower
from statsmodels.stats.proportion import proportion_effectsize
import numpy as np
```

```python
import math
def calculate_sample_size(variable_type, control_mean_or_rate=None, experiment_mean_or_rate=None, std=None,
                           power=0.8, alpha=0.05):
    if variable_type == "이산형":
        # 효과 크기 계산
        effect_size = proportion_effectsize(control_mean_or_rate, experiment_mean_or_rate)
        # 샘플 사이즈 계산
        sample_size_per_group = NormalIndPower().solve_power(
            effect_size=effect_size,
            alpha=alpha,
            power=power,
            ratio=1.0,
            alternative='two-sided'
        )
    elif variable_type == "연속형":
        # 효과 크기 계산
        effect_size = (experiment_mean_or_rate - control_mean_or_rate) / std
        # 샘플 사이즈 계산
        sample_size_per_group = TTestIndPower().solve_power(
            effect_size=effect_size,
            alpha=alpha,
            power=power,
            ratio=1.0,
            alternative='two-sided'
        )
    
    # 실험군별 사이즈와 전체 사이즈 반환
    return {
        "실험군별 샘플 사이즈": math.ceil(sample_size_per_group),
        "총 샘플 사이즈": math.ceil(sample_size_per_group * 2)  # 두 그룹의 총 사이즈
    }


inputs = [
    gr.Radio(choices=["이산형", "연속형"], label="목표지표 타입", value="이산형"),
    gr.Number(label="대조군 비율 또는 평균"),
    gr.Number(label="실험군 비율 또는 평균"),
    gr.Number(label="표준편차"),
    gr.Number(label="검정력", value=0.8),
    gr.Number(label="유의수준", value=0.05)
]

iface = gr.Interface(
    fn=calculate_sample_size,
    inputs=inputs,
    outputs="json",
    
    title="A/B 테스트를 위한 필요 샘플 사이즈 계산기",
    description="변수 유형을 선택하고 필요한 지표를 입력하여 샘플 크기를 계산합니다."
)

# 서버 실행
iface.launch(server_name="0.0.0.0", server_port=7860, height=900)
```


### 1종 오류와 2종 오류
___
* 1종 오류와 2종 오류는 통계적 가설 검정에서 발생하는 두 가지 주요 오류.

* 1종 오류 (Type I error)
    - 정의: 참인 귀무 가설을 잘못 기각하는 오류

    - 예시: 실제로 효과가 없음에도 불구하고 효과가 있다고 잘못 판단하는 경우.
    - 유의 수준 (α): 1종 오류를 범할 확률로, 일반적으로 0.05 (5%)로 설정됨
    - __1종 오류를 신경써야하는 사례__ (비용을 낭비하거나, 고객 불만족으로 이어지는 경우들)
        - 온라인 광고 캠페인: 기업이 새로운 광고 전략을 테스트하고 있는데, 효과가 없는 전략을 효과적이라고 잘못 결론 내리면, 비용과 자원의 낭비로 이어질 수 있음. 또한 장기적으로는 브랜드 가치나 시장 점유율에 부정적 영향을 줄 수 있어, 잘못된 전략을 중단하고 정확한 결론을 내리는 것이 중요함.

        - 웹사이트 디자인 변경: 웹사이트의 사용자 경험을 개선하기 위해 새로운 디자인을 도입했을 때, 실제로는 효과가 없거나 부정적인 영향을 미치는 디자인을 성공적이라고 판단하면 사용자 만족도 감소와 이탈률 증가로 이어질 수 있음.

* 2종 오류 (Type II error)
    - 정의: 거짓인 귀무 가설을 잘못 채택하는 오류

    - 예시: 실제로 효과가 있음에도 불구하고 효과가 없다고 잘못 판단하는 경우.
    - 검정력 (1 - β): 2종 오류를 범하지 않을 확률로, 일반적으로 0.8 (80%)을 목표
    - __2종 오류를 신경써야하는 사례__ (시장 기회를 놓치는 경우들)
        - 이메일 마케팅 전략: 새로운 이메일 마케팅 캠페인을 평가할 때, 실제로는 효과가 있는 캠페인을 효과가 없다고 잘못 판단하면, 매출 증대와 고객 관계 강화라는 기회를 놓칠 수 있음. 따라서 캠페인의 잠재력을 놓치지 않도록 2종 오류를 줄이는 것이 중요함.

        - 제품 가격 조정: 가격 조정이 실제로 판매량이나 고객 만족도에 긍정적인 영향을 미칠 수 있는데, 그 효과를 감지하지 못하고 기존의 가격 정책을 유지하는 경우, 추가 수익이나 시장 점유율 확대 기회를 잃을 수 있음.

<mark>__1종 오류와 2종 오류 사이에는 상충 관계가 존재하여, 1종 오류의 확률을 낮추면 2종 오류의 확률이 높아지고, 반대로 2종 오류의 확률을 낮추면 1종 오류의 확률이 높아짐. 따라서 통계적 검정을 설계할 때는 이 두 오류의 균형을 고려해야함.__</mark>


* <mark>여기서 잠깐, scipy 분포의 주요 메소드들을 살펴봅시다</mark>

   - ppf(percent point function): 분포의 누적 분포 함수(CDF)의 역함수입니다. 주어진 확률에 대한 통계량의 경계를 반환합니다.

   - pmf(probability mass function): 이산 확률 분포에서 특정 통계량의 확률을 반환합니다.
   - cdf(cumulative distribution function): 주어진 통계량까지의 확률을 누적하여 반환합니다.
   - sf(survival function): 주어진 통계량보다 큰 값의 누적 확률을 반환합니다.
   - isf(inverse survival function): 주어진 확률에 대한 통계량의 상한을 반환합니다.

```python
import plotly.graph_objects as go
import numpy as np
from scipy.stats import norm

# 정규분포 설정
mu_null = 0  # 귀무가설 하의 평균
mu_alt = 1   # 대립가설 하의 평균
sigma = 1    # 표준 편차
x = np.linspace(-3, 4, 1000)

# 귀무 가설과 대립 가설 하의 정규분포 PDF 계산
pdf_null = norm.pdf(x, mu_null, sigma)
pdf_alt = norm.pdf(x, mu_alt, sigma)

# Plotly 시각화
fig = go.Figure()

# 귀무가설 분포
fig.add_trace(go.Scatter(x=x, y=pdf_null, mode='lines', name='귀무 가설 분포 (A그룹, H0)', line=dict(color='blue')))

# 대립가설 분포
fig.add_trace(go.Scatter(x=x, y=pdf_alt, mode='lines', name='대립 가설 분포 (B그룹, H1)', line=dict(color='red')))

# 임계값 및 오류 영역 설정
alpha = 0.05
z_alpha = norm.ppf(1 - alpha, mu_null, sigma)  # 유의 수준에 따른 임계값
x_alpha = np.linspace(z_alpha, 4, 1000)
x_beta = np.linspace(-3, z_alpha, 1000)

# 1종 오류 영역 표시
fig.add_trace(go.Scatter(x=x_alpha, y=norm.pdf(x_alpha, mu_null, sigma), fill='tozeroy', name='1종 오류 (α) 영역', line=dict(color='blue'), fillcolor='rgba(0, 0, 255, 0.3)'))

# 2종 오류 영역 표시
fig.add_trace(go.Scatter(x=x_beta, y=norm.pdf(x_beta, mu_alt, sigma), fill='tozeroy', name='2종 오류 (β) 영역', line=dict(color='red'), fillcolor='rgba(255, 0, 0, 0.3)'))

# 추가 시각화 설정
fig.update_layout(title='1종 오류와 2종 오류(MDE가 낮을때)',
                  xaxis_title='테스트 통계량',
                  yaxis_title='확률 밀도',
                  legend_title="범례")

fig.show()
```

### p-value
___
* 데이터가 귀무 가설을 지지하는 증거의 강도를 나타냄. 

* p-value가 작을수록, 우리가 관찰한 데이터가 귀무 가설 하에서 발생할 가능성이 낮음을 의미함. 
* 일반적으로 p-value가 작다는 것은 관찰된 효과가 우연이 아니라 실제 차이 때문일 가능성이 높음을 나타냄.

### 단측검정과 양측검정, 그리고 비열등성 테스트
___
* 단측검정
    - 정의 : 기대하는 대안 가설의 목표 메트릭이 방향성을 가지면서 유의한 차이를 나타내는지 검증하고자 할때 수행하는 테스트
    - 예시 : 로그인 수단에 카카오 소셜로그인을 추가하면 기존 로그인 방식보다 회원가입률이 증가할 것이라는 가설을 테스트하고 싶을 때
    - 가설
        - 귀무가설(H0): 카카오 소셜로그인이 추가된 로그인의 회원가입율 ≤ 기존 로그인 수단의 회원가입율
        - 대립가설(H1): 카카오 소셜로그인이 추가된 로그인의 회원가입율 > 기존 로그인 수단의 회원가입율
    - 활용 :
        - 샘플사이즈 확보의 제한을 받는 상황이거나, 각 트리트먼트 효과의 방향성에 대한 확신을 가지고 있는 상황
        - __비열등성 테스트__ 로 활용됨.
    - 한계
        - 대안가설의 우월성만 테스트 할수 있고, 열등성을 보이는지 확인할수 없어, 비즈니스적 리스크를 안고 있음

* 양측검정
    - 정의 : 기대하는 대안 가설이 대립 가설의 메트릭과 유의한 차이가 있는자만 테스트

    - 예시 : 기존 디자인에서 새로운 웹페이지 디자인이 클릭률에 미치는 긍부정 변화를 모두 확인하고 싶을 때
    - 가설
        - 귀무가설(H0): 새 디자인의 클릭률 = 기존 디자인의 클릭률
        - 대립가설(H1): 새 디자인의 클릭률 ≠ 기존 디자인의 클릭률
    - 활용
        - 단측 검정을 하기 전에 사전 검정의 성격으로 활용됨.

    - 한계
        - 차이의 방향 불분명: 양측 검정은 두 집단 사이에 차이가 있다는 것만을 보여줌. 어느 집단이 더 우월한지, 또는 불리한지에 대한 정보는 모름
        - 결과 해석: t-통계량의 부호(+/-)를 확인하여 어느 그룹이 평균적으로 더 큰 값을 갖는지 알 수는 있지만, 이것만으로는 "우월성"을 명확하게 주장어려움.


* 비열등성 테스트
    - 비열등성 테스트는 새로운 치료법이 기존의 치료법에 비해 통계적으로 유의미하게 열등하지 않음을 증명하기 위해 사용됨. 

    - 비열등성 테스트의 샘플 사이즈는 비열등성 마진과 해당 실험에서 필요로 하는 검정력과 유의 수준을 기반으로 계산됨.
    - ( 마진이란? 마진 아래로 차이나는 것은 차이가 없다고 간주함. )
    - 비열등성 테스트는 특히 트리트먼트가 기존보다 열등하지 않다는 것을 증명해야 하므로, 이를 증명하기 위해 필요한 증거의 수준(즉, 통계적 검정력)이 매우 중요. 따라서 비교적 큰 샘플 사이즈가 필요.
    - 검정력이 충분하지 않으면, 실제로 차이가 있음에도 불구하고 이를 감지하지 못할 위험이 발생함.
    - 통계적 구조  
        | 요소 | 설명 |
        |------|------|
        | 귀무가설(H₀) | 새로운 것이 더 나쁘다 (효과 차이 ≤ −Δ) |
        | 대립가설(H₁) | 새로운 것이 비열등하다 (효과 차이 > −Δ) |
        | Δ: 비열등성 마진 | 허용 가능한 최대 차이. 예: -5%, -0.1 등 |
        | 검정 방식 | → 오른쪽 한쪽 꼬리만 보는 단측 검정입니다. |
        


    - 사례
        - 특정 UI의 복잡성을 낮추기 위해, 디자인 요소를 덜어내는 작업을 한뒤, 비즈니스 메트릭에 유해한 영향이 없는지 판단.
        - 특정 서비스를 유지하기 위한 비용을 절감하기 위해, 일부 서비스를 중단해도 비즈니스에 악영향이 없는지 판단하고 싶을때.

    - 장점: 테스트 설계 비용이 비용이 거의 들어가지 않아, 자주 테스트가 가능함.

* 요약비교
    
    | 테스트 유형 | 귀무가설 | 대립가설 | 꼬리 방향 |
    |------------|----------|----------|------------|
    | 양측 검정 | 차이 = 0 | 차이 ≠ 0 | 양측 |
    | 단측 검정 | 차이 ≤ 0 or ≥ 0 | 차이 > 0 or < 0 | 한쪽 |
    | 비열등성 검정 | 차이 ≤ −Δ | 차이 > −Δ | 오른쪽 단측 |
    | 동등성 검정 | 차이 ≥ Δ | 차이 < Δ | 양측 |



* <mark>양측검정, 단측검정, 비열등성 테스트의 샘플사이즈 계산방법의 차이</mark>

```python
from statsmodels.stats.power import TTestIndPower

# 샘플 사이즈 계산을 위한 설정
effect_size = 0.5  # 예상되는 효과 크기
alpha = 0.05       # 유의 수준
power = 0.8        # 검정력

# 양측검정의 샘플 사이즈 계산
power_analysis = TTestIndPower()
sample_size_two_sided = power_analysis.solve_power(effect_size=effect_size, alpha=alpha, power=power, alternative='two-sided')
print(f'양측검정 필요 샘플 사이즈: {sample_size_two_sided:.2f}')

# 단측검정의 샘플 사이즈 계산
sample_size_one_sided = power_analysis.solve_power(effect_size=effect_size, alpha=alpha, power=power, alternative='larger')
print(f'단측검정 필요 샘플 사이즈: {sample_size_one_sided:.2f}')

# 비열등성 테스트의 샘플 사이즈 계산
# 비열등성 마진을 반영한 효과 크기 조정 (여기서는 단순 예시로 작은 값 사용)
non_inferiority_margin = 0.1
effect_size_non_inferiority = effect_size - non_inferiority_margin
sample_size_non_inferiority = power_analysis.solve_power(effect_size=effect_size_non_inferiority, alpha=alpha, power=power, alternative='two-sided')
print(f'비열등성 테스트 필요 샘플 사이즈: {sample_size_non_inferiority:.2f}')

```

## 5. 가상 데이터로 ABTEST 해보기
___

* 20일간 20000명의 사용자가 평균 3번의 방문을 한다고 가정한 방문데이터 제네레이터

```python
# %run generate_data.py
```

```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind, t
from scipy import stats

def calculate_confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    stderr = stats.sem(data)
    interval = stderr * t.ppf((1 + confidence) / 2., n-1)
    return (mean - interval, mean + interval)

# 데이터 파일 로드
file_path = "simulation/simulation0.csv"
data = pd.read_csv(file_path)
data.head()
```

```python
# 'variant' 열을 기준으로 데이터 분리
group_A = data[data['variant'] == 'A']
group_B = data[data['variant'] == 'B']

# 'normal_shifted', 'cr_shifted' 지표를 사용하여 두 그룹 간의 t-검정 수행
t_stat, p_value = ttest_ind(group_A['normal_shifted'], group_B['normal_shifted'], equal_var=False)
t_stat_cr, p_value_cr =  ttest_ind(group_A['cr_shifted'], group_B['cr_shifted'], equal_var=False)

# 신뢰구간 계산
ci_A_normal = calculate_confidence_interval(group_A['normal_shifted'])
ci_B_normal = calculate_confidence_interval(group_B['normal_shifted'])

# 전환율('cr_shifted')에 대한 신뢰구간 계산
ci_A_cr = calculate_confidence_interval(group_A['cr_shifted'])
ci_B_cr = calculate_confidence_interval(group_B['cr_shifted'])

print(f"그룹 A 'normal_shifted' 신뢰구간: {ci_A_normal}")
print(f"그룹 B 'normal_shifted' 신뢰구간: {ci_B_normal}")
print(f"그룹 A 'cr_shifted' 신뢰구간: {ci_A_cr}")
print(f"그룹 B 'cr_shifted' 신뢰구간: {ci_B_cr}")

# 결과 해석
if p_value < 0.05:
    print(f"'normal_shifted'에서 그룹 A와 B는 통계적 유의미한 차이 존재 (p-val : {p_value})")
else:
    print(f"'normal_shifted'에서 그룹 A와 B는 통계적 유의미한 차이 없음 (p-val : {p_value})")

if p_value_cr < 0.05:
    print(f"'cr_shifted'에서 그룹 A와 B는 통계적 유의미한 차이 존재 (p-val : {p_value_cr})")
else:
    print(f"'cr_shifted'에서 그룹 A와 B는 통계적 유의미한 차이 없음 (p-val : {p_value})")
```
